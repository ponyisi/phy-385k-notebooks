```python
import numpy
from matplotlib import pyplot as plt
```

```python
ARNG = numpy.linspace(1, 4, 1000)
INIT = numpy.array([0.5]*len(ARNG), 'f')

# total number of iterations N and number to ignore at beginning CUTOFF
N = 1000
CUTOFF = 500

# logisitic map
def logstep(a, x):
    return a*x*(1-x)

# derivative of map, used to calculate Lyuapunov exponents later
def logderiv(a, x):
    return a*(1-2*x)

arrays = []
x = INIT
for i in range(N):
    x = logstep(ARNG, x)
    if i >= CUTOFF:
        arrays.append(x)

a = numpy.stack(arrays)
a2 = numpy.concatenate(arrays)
rng2 = numpy.tile(ARNG, (N-CUTOFF))
```

```python
plt.scatter(rng2,a2,s=0.1)
plt.show()
```

```python
ARNG = numpy.linspace(1, 4, 1000)
INIT = numpy.array([0.5]*len(ARNG), 'f')
DELTA = 0.001
INIT2 = INIT + DELTA

arrays = []
x = INIT
for i in range(N):
    x = logstep(ARNG, x)
    if i >= CUTOFF:
        arrays.append(x)

x = arrays[0]+DELTA
arraysdelta = []
for i in range(N):
    x = logstep(ARNG, x)
    if i >= CUTOFF:
        arraysdelta.append(x)
```

```python
plt.scatter(ARNG, arrays[-1], s=0.1)
plt.scatter(ARNG, arraysdelta[-2], s=0.1)
```

```python
ARNG = numpy.linspace(1.0000, 3.999999, 4000)
INIT = numpy.array([0.5]*len(ARNG), 'f')

derivs = []
x = INIT
for i in range(N):
    x = logstep(ARNG, x)
    if i >= CUTOFF:
        derivs.append(numpy.log(numpy.abs(logderiv(ARNG,x))))

plt.scatter(ARNG, numpy.mean(derivs, axis=0), s=0.4)
plt.show()
```
